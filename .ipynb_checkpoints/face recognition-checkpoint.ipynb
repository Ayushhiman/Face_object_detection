```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
import cv2
import os
from fr_utils import *
from inception_blocks_v2 import *
from yolo_utils import read_classes, read_anchors, generate_colors, preprocess_image, draw_boxes, scale_boxes
from yad2k.models.keras_yolo import yolo_head, yolo_boxes_to_corners, preprocess_true_boxes, yolo_loss, yolo_body
keras.backend.set_image_data_format('channels_first')
%matplotlib inline
```

```python
FRmodel = faceRecoModel(input_shape=(3, 96, 96))
load_weights_from_FaceNet(FRmodel)
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def img_to_face_encoding(img, model):
    img = img[...,::-1]
    img = np.around(np.transpose(img, (2,0,1))/255.0, decimals=12)
    x_train = np.array([img])
    encoding = model.predict_on_batch(x_train)
    return encoding
```

```python
def histogram(img):
    b,g,r = cv2.split(img)
    equ_b = cv2.equalizeHist(b)
    equ_g = cv2.equalizeHist(g)
    equ_r = cv2.equalizeHist(r)
    equ = cv2.merge((equ_b, equ_g, equ_r))
    return equ
```

```python
def data_to_encoding(path,model):
    img = cv2.imread(path)
    img=cv2.resize(img,(600,400))
    img=cv2.flip(img,1)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.1, 10)
    x,y,w,h=faces[0]
    crop_image=cv2.resize(img[y:y+h,x:x+w,:], (96, 96)) 
    encoding=img_to_face_encoding(crop_image,model)
    return encoding
```

```python
def find_face(img,x,y,w,h,database,model):
    crop_image=cv2.resize(img[y:y+h,x:x+w,:], (96, 96)) 
    encoding=img_to_face_encoding(crop_image,model)
    min_dist = 100
    for (name, db_enc) in database.items():
        dist = np.linalg.norm(encoding - db_enc)
#         print(dist)
        if dist<min_dist:
            min_dist = dist
            identity = name
    if min_dist > 1:
        return 'unknown'
    else:
        return identity
```

```python
database = {}
database['ayush']=data_to_encoding("Ayush.jpg",FRmodel)
```

```python
class_names = read_classes("model_data/coco_classes.txt")
anchors = read_anchors("model_data/yolo_anchors.txt")
```

```python
yolo_model = keras.models.load_model("model_data/yolo.h5")
```

```python
def yolo_filter_boxes(box_confidence, boxes, box_class_probs, threshold = .6):
    box_scores = box_confidence * box_class_probs
    box_classes = K.argmax(box_scores, axis=-1)
    box_class_scores = K.max(box_scores, axis=-1)
    filtering_mask = (box_class_scores >= threshold)
    scores = tf.boolean_mask(box_class_scores, filtering_mask)
    boxes = tf.boolean_mask(boxes, filtering_mask)
    classes = tf.boolean_mask(box_classes, filtering_mask)
    
    return scores, boxes, classes
```

```python
def yolo_prediction(input_img,img_equ,model):
    font = cv2.FONT_HERSHEY_SIMPLEX 
    image_shape=(float(input_img.shape[0]),float(input_img.shape[1]))
    img=cv2.resize(img_equ,(608,608))
    img = img[...,::-1]
    img = np.around(img/255.0, decimals=12)
    x_train = np.array([img])
    yolo_outputs = yolo_head(model.output, anchors, len(class_names))
    box_confidence, box_xy, box_wh, box_class_probs = yolo_outputs
    boxes = yolo_boxes_to_corners(box_xy, box_wh)
    scores, boxes, classes = yolo_filter_boxes(box_confidence, boxes, box_class_probs, 0.6)
    boxes = scale_boxes(boxes, image_shape)
    nms_indices = tf.image.non_max_suppression(boxes,scores,10,0.5)
    scores = K.gather(scores,nms_indices)
    boxes = K.gather(boxes,nms_indices)
    classes = K.gather(classes,nms_indices)
    out_scores, out_boxes, out_classes = K.get_session().run([scores, boxes, classes], feed_dict={model.input: x_train, K.learning_phase(): 0})
    for i, c in reversed(list(enumerate(out_classes))):
        predicted_class = class_names[c]
        if predicted_class=='person':
            continue
        box = out_boxes[i]
        score = out_scores[i]
        top, left, bottom, right = box
        label = '{} {:.2f}'.format(predicted_class, score)
        cv2.rectangle(input_img, (left,top), (right,bottom), (255, 0, 0), 1)
#         scale = ((right-left) * (top-bottom)) / (250* 250)
        cv2.putText(input_img,label,(left,top), font, 0.5,(0,0,255),2)
    return input_img
```

```python
def object_face_recognition(img,database,model1,model2):
    img=cv2.resize(img,(600,400))
    img=cv2.flip(img,1)
    img_equ=histogram(img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray= cv2.equalizeHist(gray)
    faces = face_cascade.detectMultiScale(gray, 1.1, 10)
    font = cv2.FONT_HERSHEY_SIMPLEX 
    for (x, y, w, h) in faces:
#         t=find_face(img_equ,x,y,w,h,database,model1)
        cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
#         cv2.putText(img,t,(x,y), font, 0.5,(0,0,255),2)
    img=yolo_prediction(img,img_equ,model2)
    return img
```

```python
img=object_face_recognition(cv2.imread("Ayush.jpg"),database,yolo_model,yolo_model)
```
